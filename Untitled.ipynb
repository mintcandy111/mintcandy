```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
import numpy as np
from datetime import datetime, timedelta

# 어제 날짜 계산
yesterday = datetime.today() - timedelta(days=1)
date_str = yesterday.strftime("%Y-%m-%d")

# 테슬라 주가 2일치 (1분봉) 로드 후 어제 필터
tsla = yf.Ticker("TSLA")
df = tsla.history(period="2d", interval="1m")
df = df[df.index.strftime("%Y-%m-%d") == date_str]

# 10분 단위로 종가만 평균 냄 (혹은 마지막 값으로도 가능: 'last')
df_10min = df['Close'].resample('10T').mean().dropna()

# numpy 배열로 변환
close_values = df_10min.values

# 극대, 극소 인덱스 찾기
max_idx = argrelextrema(close_values, np.greater)[0]
min_idx = argrelextrema(close_values, np.less)[0]

# 인덱스를 시계열로 변환
max_times = df_10min.iloc[max_idx].index
min_times = df_10min.iloc[min_idx].index

plt.figure(figsize=(14, 6))
plt.plot(df_10min.index, df_10min.values, label='TSLA (10-min Avg Close)', color='blue')

# 극대값 (빨간 점)
plt.scatter(max_times, df_10min.loc[max_times], color='red', label='Maxima', marker='^', s=100)

# 극소값 (초록 점)
plt.scatter(min_times, df_10min.loc[min_times], color='green', label='Minima', marker='v', s=100)

plt.title(f"TSLA 10-Minute Close Price on {date_str}")
plt.xlabel("Time")
plt.ylabel("Close Price (USD)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# 데이터 불러오기 (최근 5일치, 1분봉 → 10분봉으로 변환)
tsla = yf.Ticker("TSLA")
df = tsla.history(period="5d", interval="1m")

# 10분 단위로 리샘플
df_10min = df.resample('10T').agg({'Close': 'mean', 'Volume': 'sum'}).dropna()

# 수익률 (종가 변화율)
df_10min["Return"] = df_10min["Close"].pct_change()

# 거래량 변화율
df_10min["Vol_Change"] = df_10min["Volume"].pct_change()

# 이동평균 대비 편차
df_10min["MA_30"] = df_10min["Close"].rolling(window=3).mean()
df_10min["Dist_MA"] = df_10min["Close"] - df_10min["MA_30"]

from scipy.signal import argrelextrema
import numpy as np

# 극대/극소 인덱스
max_idx = argrelextrema(df_10min['Close'].values, np.greater, order=2)[0]
min_idx = argrelextrema(df_10min['Close'].values, np.less, order=2)[0]

# 라벨 컬럼 추가 (0: 중간, 1: 고점, -1: 저점)
df_10min["Label"] = 0
df_10min.iloc[max_idx, df_10min.columns.get_loc("Label")] = 1
df_10min.iloc[min_idx, df_10min.columns.get_loc("Label")] = -1

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# 결측 제거
df_ml = df_10min.dropna()

# 피처와 라벨 분리
X = df_ml[["Return", "Vol_Change", "Dist_MA"]]
y = df_ml["Label"]

# 학습/테스트 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 모델 훈련
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

import yfinance as yf
from datetime import datetime

# 오늘 날짜
today_str = datetime.today().strftime("%Y-%m-%d")

# 오늘 1일치 1분봉 받아서 10분봉으로 변환
tsla = yf.Ticker("TSLA")
df_today = tsla.history(period="1d", interval="1m")
df_today = df_today.resample("10T").agg({'Close': 'mean', 'Volume': 'sum'}).dropna()

df_today["Return"] = df_today["Close"].pct_change()
df_today["Vol_Change"] = df_today["Volume"].pct_change()
df_today["MA_30"] = df_today["Close"].rolling(window=3).mean()
df_today["Dist_MA"] = df_today["Close"] - df_today["MA_30"]

# 결측 제거 후 마지막 1개 샘플 사용
latest = df_today.dropna().iloc[-1]
X_now = latest[["Return", "Vol_Change", "Dist_MA"]].values.reshape(1, -1)

# 예측 수행
prediction = clf.predict(X_now)[0]

# 해석
label_map = {1: "🔺 극대 (Peak)", -1: "🔻 극소 (Valley)", 0: "⏸️ 중간 (Normal)"}
print(f"현재 시점 예측: {label_map[prediction]}")


```

```python
df_ml["Future_Return"] = df_ml["Close"].shift(-3) / df_ml["Close"] - 1  # 예: 3개 이후 수익률

df_reg = df_ml.dropna(subset=["Return", "Vol_Change", "Dist_MA", "Future_Return"])

X_reg = df_reg[["Return", "Vol_Change", "Dist_MA"]]
y_reg = df_reg["Future_Return"]

from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()
reg.fit(X_reg, y_reg)

# 최신 시점의 피처 계산
latest = df_today.dropna().iloc[-1]
X_now = latest[["Return", "Vol_Change", "Dist_MA"]].values.reshape(1, -1)

# 분류 예측
pred_class = clf.predict(X_now)[0]
label_map = {1: "🔺 극대", -1: "🔻 극소", 0: "⏸️ 중간"}

# 회귀 예측 (중간일 때만)
if pred_class == 0:
    pred_return = reg.predict(X_now)[0]
    direction = "📈 상승" if pred_return > 0 else "📉 하락"
    percent = round(pred_return * 100, 2)
    print(f"현재 상태: {label_map[pred_class]} / 앞으로 {percent}% {direction} 예상")
else:
    print(f"현재 상태: {label_map[pred_class]} (수익률 예측 생략)")
```
